# Estimating Covid19 Infected and Fatal using Age Distribution

# Loading Preliminaries

```python
import pandas as pd
import geopandas
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set(style="whitegrid")
```

# 2015 Census Data - Age Distribution per Province

```python
raw=pd.read_excel('data/Age-Distribution-per-region.xlsx', header=4)
```

```python
new=raw[1:19].groupby(raw[1:19].index // 2).sum()
```

```python
range_=['Under 1','1-9','10-19','20-29','30-39','40-49','50-59','60-69','70-79','Above 80']
```

```python
age=pd.concat([pd.DataFrame(range_, columns=['range']),new],axis=1)
```

```python
age.set_index('range')
```

### Infection using R0 (2,3,4)

```python
Infected = {R0:(1-(1/R0)) * age.set_index('range') for R0 in range(2,5)}
Infected_per_Region = Infected[3].sum(axis=0).drop('PHILIPPINES').reset_index().rename(columns={'index':'Region', 0:'Infected_3'})
```

```python
Infected_per_Region
```

### Infected per Region (R0 = 3)

```python
fig, ax = plt.subplots(figsize=(20,10))

chart = sns.barplot(x = "Region", y = "Infected_3", ax=ax, data=Infected_per_Region)
chart.set_xticklabels(
    chart.get_xticklabels(), 
    rotation=45, 
    horizontalalignment='right',
    fontweight='light',
    fontsize='medium')

chart.set_title('Infected per Region (R0 = 3)')
plt.show()
```

```python
Infected_per_Region = pd.concat([Infected_per_Region,Infected[2].sum(axis=0).drop('PHILIPPINES').reset_index().rename(columns={'index':'Region_2', 0:'Infected_2'})], axis = 1)
Infected_per_Region = pd.concat([Infected_per_Region,Infected[4].sum(axis=0).drop('PHILIPPINES').reset_index().rename(columns={'index':'Region_4', 0:'Infected_4'})], axis = 1)
Infected_per_Region.drop(['Region_2', 'Region_4'], axis =1, inplace = True)
Infected_per_Region =Infected_per_Region[['Region','Infected_2', 'Infected_3', 'Infected_4']]
```

```python
Infected_per_Region.set_index('Region')
```

### Infected per Region for different R0

```python

fig, ax = plt.subplots(3,1,figsize=(20,10))

for i in range(3):
    chart = sns.barplot(x = "Region", y = "Infected_"+str(i+2), ax=ax[i], data=Infected_per_Region)
    chart.set_xticklabels(
        chart.get_xticklabels(), 
        rotation=45, 
        horizontalalignment='right',
        fontweight='light',
        fontsize='medium')
    chart.set_title('Infected per Region R0 = '+str(i+2))
    if i !=2:
        chart.set_xticks([])
        chart.set_xlabel('')
    

plt.show()
```

### Fatality Rate, based on China numbers

<img src="images/risk rates.png">

```python
Fatality=pd.Series([0,0,0.002,0.002,0.002,0.004,0.013,0.036,0.08,0.148])
Fatal = {i:Infected[i].apply(lambda x: np.asarray(x) * np.asarray(Fatality)) for i in range(2,5)}
```

```python
regions= list(Fatal[2].drop('PHILIPPINES', axis=1).columns)
```

```python
Fatal_per_Region = pd.DataFrame({'Fatal_R'+str(i):list(Fatal[i].sum(axis=0).drop('PHILIPPINES').reset_index()[0]) for i in range(2,5)})
Fatal_per_Region['Region'] = regions
```

```python
Fatal_per_Region.set_index('Region')
```

### Fatal per Region for different R0

```python
fig, ax = plt.subplots(3,1,figsize=(20,10))

for i in range(3):
    chart = sns.barplot(x = "Region", y = "Fatal_R"+str(i+2), ax=ax[i], data=Fatal_per_Region)
    chart.set_xticklabels(
        chart.get_xticklabels(), 
        rotation=45, 
        horizontalalignment='right',
        fontweight='light',
        fontsize='medium')
    chart.set_title('Fatal per Region R0 = '+str(i+2))
    if i !=2:
        chart.set_xticks([])
        chart.set_xlabel('')
    

plt.show()
```

### Fatality Normalized by Population of Region

```python
total_pop = raw.iloc[0][2:]
```

```python
Fatal_per_Region_norm = Fatal_per_Region.set_index('Region').div(total_pop, axis=0)
Fatal_per_Region_norm
```

```python
fig, ax = plt.subplots(3,1,figsize=(20,10))

for i in range(3):
    chart = sns.barplot(x = "Region", y = "Fatal_R"+str(i+2), ax=ax[i], data=Fatal_per_Region_norm.reset_index())
    chart.set_xticklabels(
        chart.get_xticklabels(), 
        rotation=45, 
        horizontalalignment='right',
        fontweight='light',
        fontsize='medium')
    chart.set_title('Fatal per Region R0 = '+str(i+2))
    if i !=2:
        chart.set_xticks([])
        chart.set_xlabel('')
    

plt.show()
```

### Factoring in Density
If people are able to utilize the whole land area of their region

```python
density=pd.read_excel('data/density.xlsx',sheet_name = 'final-infected')
density = list(density.iloc[4][3:])
```

```python
Fatal_per_Region_dense = Fatal_per_Region.set_index('Region').mul(density, axis=0)
```

```python
Fatal_per_Region_dense
```

```python
fig, ax = plt.subplots(3,1,figsize=(20,10))

for i in range(3):
    chart = sns.barplot(x = "Region", y = "Fatal_R"+str(i+2), ax=ax[i], data=Fatal_per_Region_dense.reset_index())
    chart.set_xticklabels(
        chart.get_xticklabels(), 
        rotation=45, 
        horizontalalignment='right',
        fontweight='light',
        fontsize='medium')
    chart.set_title('Fatal per Region R0 = '+str(i+2))
    if i !=2:
        chart.set_xticks([])
        chart.set_xlabel('')
    
plt.show()
```

# Geomap of # of at risk vs number of cases

```python
import geopandas
import json
```

```python
raw_geo = geopandas.read_file("data/gadm36_PHL.gpkg")
# download shapefiles from https://gadm.org/download_country_v3.html
```

```python
with open("data/philippine_provinces_cities_municipalities_and_barangays_2019v2.json") as myjsonfile:
    ph_data = json.load(myjsonfile)  
```

```python
province_per_region = {ph_data[i]['region_name']:list(ph_data[i]['province_list'].keys()) for i in ph_data.keys()}
```

```python
Infected_per_Region.set_index('Region', inplace = True)
```

```python
Infected_per_Region.rename({i:i.split('-')[0].strip() for i in Infected_per_Region.index}, inplace=True)
```

```python
Infected_per_Region.rename({'NATIONAL CAPITAL REGION':'NCR', 'CORDILLERA ADMINISTRATIVE REGION':'CAR', 'MIMAROPA REGION':'REGION IV-B',
                        'REGION IV':'REGION IV-A','AUTONOMOUS REGION IN MUSLIM MINDANAO':'BARMM'}, inplace = True)
```

```python
geo_per_province = raw_geo[['GID_1', 'NAME_1', 'geometry']]
```

```python
province_per_region['REGION XI'].append('DAVAO DEL NORTE')
province_per_region['REGION VIII'].append('SAMAR')
province_per_region['NCR'].append('METROPOLITAN MANILA')
province_per_region['REGION XII'].append('NORTH COTABATO')
```

```python
region=[]
for i in geo_per_province.NAME_1:
    match = 0
    for j,k in province_per_region.items():
        if i.upper() in k:
            region.append(j)
            match = 1
    
    if match == 0:
        region.append(None)
            
```

```python
geo_per_province['Region'] = region
```

```python
final_infected_region = pd.merge(geo_per_province,Infected_per_Region, on='Region', how='left')
```

```python
final_infected_region = final_infected_region.dissolve(by='Region', aggfunc='first').reset_index()
# collapsed per region
```

```python
final_infected_region.to_file("final_infected_region.json", driver="GeoJSON")
```
